```python
driver_path = "/opt/homebrew/bin/chromedriver"

year = kbodata.get_yearly_schedule(2025, driver_path)
```

```python
year_data = kbodata.get_game_data(year, driver_path)
```

```python
print(year_data[0])
```

```python
import json

with open("year_data_2025.json", "w", encoding="utf-8") as f:
    json.dump(year_data, f, ensure_ascii=False, indent=2)
```

```python
scoreboard = kbodata.scoreboard_to_DataFrame(year_data)
    # 타자 정보만을 정리하여 DataFrame으로 변환합니다.
batter = kbodata.batter_to_DataFrame(year_data)
    # 투수 정보만을 정리하여 DataFrame으로 변환합니다.
pitcher = kbodata.pitcher_to_DataFrame(year_data)
```

```python
import os

# 저장 디렉터리 (선택)
os.makedirs("csv", exist_ok=True)

scoreboard.to_csv(
    "csv/scoreboard_2025.csv",
    index=False,
    encoding="utf-8-sig"
)

batter.to_csv(
    "csv/batter_2025.csv",
    index=False,
    encoding="utf-8-sig"
)

pitcher.to_csv(
    "csv/pitcher_2025.csv",
    index=False,
    encoding="utf-8-sig"
)
```

```python
import sys, subprocess, shlex, textwrap
cmd = [sys.executable, '-m', 'pip', 'install', '--no-cache-dir', 'kbodata']
print('Running:', ' '.join(shlex.quote(c) for c in cmd))
proc = subprocess.run(cmd, check=False, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
print('Return code:', proc.returncode)
print('--- output ---')
print('\n'.join(proc.stdout.splitlines()[-20:]))
```

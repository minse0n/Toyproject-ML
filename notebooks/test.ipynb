```python
import kbodata
import time
import pandas as pd

# 1. 스케줄 가져오기
month_schedule = kbodata.get_monthly_schedule(2024, 8, driver_path)

month_data_list = []

print(f"총 {len(month_schedule)}개의 경기를 수집합니다.")

# 2. 수집 루프
for i in range(len(month_schedule)):
    try:
        # 데이터프레임 형태 유지하며 한 줄 추출
        single_row_df = month_schedule.iloc[[i]]
        print(f"번호 {i} 경기 데이터 수집 중...")
        # 실제 데이터 가져오기
        single_game_data = kbodata.get_game_data(single_row_df, driver_path)
        month_data_list.append(single_game_data)
        
        time.sleep(1.2) # 조금 더 여유 있게 대기
        
    except Exception as e:
        print(f"번호 {i}에서 오류 발생: {e}")
        continue

print(f"수집 완료! 총 {len(month_data_list)}개의 경기 데이터를 모았습니다.")
```

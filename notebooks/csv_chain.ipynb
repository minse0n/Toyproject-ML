```python
import os
import pandas as pd

# "csv/scoreboard_2024_06_2025.csv"는 2024년 6월 - 2025년 전체 기록
# "csv/pitcher_2024_06_2025.csv"는 2024년 6월 - 2025년 전체 기록
# "csv/batter_2024_06_2025.csv"는 2024년 6월 - 2025년 전체 기록
# "csv/scoreboard_2024_{month}.csv"는 {month}에 해당하는 달의 스코어보드 기록입니다.
# "csv/pitcher_2024_{month}.csv"는P{month}에 해당하는 달의 투수 기록입니다.
# "csv/batter_2024_{month}.csv"는P{month}에 해당하는 달의 타자 기록입니다.
# "csv/pitcher_2023_year.csv"는 2023년 전체 기록
# "csv/batter_2023_year.csv"는 2023년 전체 기록
# "csv/scoreboard_2023_year.csv"는 2023년 전체 기록
#  month에 들어갈건 march, april, may
# 모든 csv는 첫번째 열은 컬럼명 입니다.
# "csv/pitcher_2023_2024_2025.csv"는 2023년, 2024년, 2025년 전체 기록을 만들고 싶음 

pitcher_2024_06_2025 = pd.read_csv("csv/pitcher_2024_06_2025.csv")
pitcher_year2023 = pd.read_csv("csv/pitcher_2023_year.csv")
pitcher_march24 = pd.read_csv("csv/pitcher_2024_march.csv")
pitcher_april24 = pd.read_csv("csv/pitcher_2024_april.csv")
pitcher_may24 = pd.read_csv("csv/pitcher_2024_may.csv")

pitcher_2024_2025 = pd.concat(
    [pitcher_march24, pitcher_april24, pitcher_may24, pitcher_2024_06_2025],
    ignore_index=True
)

pitcher_2023_2024_2025 = pd.concat(
    [pitcher_year2023, pitcher_2024_2025],
    ignore_index=True
)

pitcher_2023_2024_2025.to_csv(
    "csv/pitcher_2023_2024_2025.csv",
    index=False
)
```

```python
batter_2024_06_2025 = pd.read_csv("csv/batter_2024_06_2025.csv")  
batter_year2023 = pd.read_csv("csv/batter_2023_year.csv")
batter_march24 = pd.read_csv("csv/batter_2024_march.csv")
batter_april24 = pd.read_csv("csv/batter_2024_april.csv")
batter_may24 = pd.read_csv("csv/batter_2024_may.csv")
batter_2024_2025 = pd.concat(
    [batter_march24, batter_april24, batter_may24, batter_2024_06_2025],
    ignore_index=True
)
batter_2023_2024_2025 = pd.concat(
    [batter_year2023, batter_2024_2025],
    ignore_index=True
)

batter_2023_2024_2025.to_csv(
    "csv/batter_2023_2024_2025.csv",
    index=False
)
```

```python
scoreboard_2024_06_2025 = pd.read_csv("csv/scoreboard_2024_06_2025.csv")  
scoreboard_year2023 = pd.read_csv("csv/scoreboard_2023_year.csv")
scoreboard_march24 = pd.read_csv("csv/scoreboard_2024_march.csv")
scoreboard_april24 = pd.read_csv("csv/scoreboard_2024_april.csv")
scoreboard_may24 = pd.read_csv("csv/scoreboard_2024_may.csv")
scoreboard_2024_2025 = pd.concat(
    [scoreboard_march24, scoreboard_april24, scoreboard_may24, scoreboard_2024_06_2025],
    ignore_index=True
)
scoreboard_2023_2024_2025 = pd.concat(
    [scoreboard_year2023, scoreboard_2024_2025],
    ignore_index=True
)

scoreboard_2023_2024_2025.to_csv(
    "csv/scoreboard_2023_2024_2025.csv",
    index=False
)
```
